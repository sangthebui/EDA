# Create a table

```python
%%sql
CREATE TABLE users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT not null,
    email TEXT UNIQUE
)
```

# Insert data

```python
%%sql
INSERT INTO users (name, email) VALUES ('Alice', 'alice@example.com');
INSERT INTO users (name, email) VALUES ('Bob', 'bob@example.com');
```

# Select data

```python
result = %sql SELECT name, email FROM users;
```
