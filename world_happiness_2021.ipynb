# World Happiness Report 2021 (EDA)

## About Dataset
### Context
- The World Happiness Report is a landmark survey of the state of global happiness. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

### Content
- The happiness scores and rankings use data from the Gallup World Poll . The columns following the happiness score estimate the extent to which each of six factors – economic production, social support, life expectancy, freedom, absence of corruption, and generosity – contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

# Question 1: What are the top 10 happiest countries?

```python
# Set the style for plots
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Question 1: What are the top 10 happiest countries?
top_10_happy = data.nlargest(10, 'Ladder score')

plt.figure(figsize=(12, 7))
sns.barplot(x='Ladder score', y='Country name', data=top_10_happy, palette='viridis')
plt.title('Top 10 Happiest Countries in 2021', fontsize=16)
plt.xlabel('Happiness Score (Ladder score)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
```

# Question 3: How do happiness scores vary by region?

```python
regional_happiness = data.groupby('Regional indicator')['Ladder score'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Ladder score', y='Regional indicator', data=regional_happiness, palette='plasma')
plt.title('Average Happiness Score by Region', fontsize=16)
plt.xlabel('Average Happiness Score', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.show()
```

# --- 3. Deeper Dive into Key Relationships ---

```python
# Let's explore the relationship between GDP and happiness score.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Logged GDP per capita', y='Ladder score', data=data, hue='Regional indicator', s=100)
plt.title('Happiness Score vs. GDP per Capita', fontsize=16)
plt.xlabel('Logged GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
# This plot clearly shows a positive linear relationship: as GDP increases, the happiness score tends to increase.

# Let's explore the relationship between Social Support and happiness score.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Social support', y='Ladder score', data=data, hue='Regional indicator', s=100, palette='magma')
plt.title('Happiness Score vs. Social Support', fontsize=16)
plt.xlabel('Social Support', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
# A strong positive relationship is also visible here.

print("--- EDA Complete ---")
print("The plots above provide insights into the World Happiness Report 2021.")
```
